# ZSCORE METHOD


```python
from scipy.stats import zscore

z= np.abs(zscore(df))
print(np.where(z>3))
```

```python
df_1 = df[(z<3).all(axis=1)]
print('With outliers::',df.shape)
print("After removing outliers::",df_1.shape)
```

zscore removes 42 rows from dataset


# IQR

```python
from scipy import stats
IQR = stats.iqr(df)
IQR
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
df_out = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_out.shape)
```

Here choosing ZSCORE because there is huge dataloss in IQR method


```python
df = df_1
```

# Imbalanced learn


Using OverSampling_BorderlineSMOTE

because there is less data in attrition for yes

```python
sns.set_style("whitegrid")
sns.countplot(x='Attrition', data=df)
```

We can see here Data is not fully balance but we'll treat the imbalance

We have around 200 in no and around 1200 in yes

```python
from imblearn.over_sampling import BorderlineSMOTE
x_rus, y_rus =BorderlineSMOTE().fit_resample(x,y)
print('original Target dataset shape:',y.shape)
print('Resample Target dataset shape',y_rus.shape)
```

```python
sns.set_style("whitegrid")
sns.countplot(y_rus)
```

So now we have balanced data for model training.


# Spliting Train and Test Data


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_rus,y_rus,test_size=.27,random_state=42)
```

# LogisticRegression

# HYPER PARAMETER TUNING


```python
adb=AdaBoostClassifier()
param_grid={
       'algorithm' : ['SAMME.R'],
       'n_estimators' : [50,200,10],
       'learning_rate' : [0.1,0.5,1.0],
      'random_state' : [10]
}
```

```python
adb_grid=GridSearchCV(AdaBoostClassifier(),param_grid,cv=4,scoring='accuracy',n_jobs=-1,verbose=2)
```

```python
adb_grid.fit(x_train,y_train)
adb_pred=adb_grid.best_estimator_.predict(x_test)
print("Accuracy after parameter tuning::",accuracy_score(y_test,adb_pred))
```

```python
adb_grid.best_params_
```

# Model training with best parameters


```python
best_param={
     'algorithm' : ['SAMME.R'],
       'n_estimators' : [50,200,10],
       'learning_rate' : [0.1,0.5,1.0],
      'random_state' : [10]
}
```

```python
best_adb_grid=GridSearchCV(AdaBoostClassifier(),best_param,cv=4,scoring='accuracy',n_jobs=-1,verbose=2)
```
